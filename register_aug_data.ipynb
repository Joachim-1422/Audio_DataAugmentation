```python
from wav_augmentation import WavAugmentation
import os
%matplotlib inline
```

First, let's use only 1 audio file and check the data augmentation.

```python
wav = WavAugmentation('./Data/Actor_01/03-01-01-01-01-01-01.wav')
print('normal data')
norm_wav = wav.wav
wav.plot_time_series(norm_wav)
print('data with noise')
wav.plot_time_series(wav.add_noise())
print('shifted data')
wav.plot_time_series(wav.shift())
print('stretched data')
wav.plot_time_series(wav.stretch())
```

Now I will make a for loop who will take all the audio files and register them as new augmented wav file into they're respective folder NoiseData, ShiftData, StretchData.

```python
for subdir, dirs, files in os.walk('./Data'):
    for file in files:
        fn = str(file)
        wav = WavAugmentation(os.path.join(subdir, file))
        wav.register_wav('NoiseData/' + fn, wav.add_noise())
        wav.register_wav('ShiftData/' + fn, wav.shift())
        wav.register_wav('StretchData/' + fn, wav.stretch(1.2))
```
